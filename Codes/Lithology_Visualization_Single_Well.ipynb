```python
import lasio
import pandas as pd
```

```python
# Load the LAS file
las = lasio.read("15_9-13.las")  # Update path if needed

# Print basic header and curve info
print(las.header)
print("\nAvailable Curves:")
for curve in las.curves:
    print(f"{curve.mnemonic} ({curve.unit}): {curve.descr}")


# Convert data to a DataFrame
df = las.df().reset_index()  # Reset index so depth becomes a column
df.head()
```

```python
df_relevant= df[['DEPTH_MD', 'GR', 'RMED', 'SP','RHOB','NPHI', 'FORCE_2020_LITHOFACIES_LITHOLOGY']]
df_relevant.isna().sum()
```

```python
df_clean = df_relevant.dropna()
print(f"Original rows: {len(df_relevant)}")
print(f"Cleaned rows: {len(df_clean)}")
df_clean.head()
```

```python
df_clean.tail()
```

```python
min_depth = 2360
max_depth = 2400

df_filtered = df_clean[(df_clean['DEPTH_MD'] >= min_depth) & (df_clean['DEPTH_MD'] <= max_depth)]
df_filtered.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patches as mpatches

# Define plotting parameters
logs = ['GR', 'RMED', 'SP', 'RHOB', 'NPHI']
colors = ['red', 'blue', 'green', 'brown', 'orange']
xlabels = ['GR (API)', 'RMED (ohm·m)', 'SP (mV)', 'RHOB (g/cm³)', 'NPHI (m³/m³)']

# Lithology color mapping for legend and display
lithology_keys = {
    30000: 'Sandstone',
    65000: 'Shale'
}
lithology_colors = {
    30000: 'orange',
    65000: 'red'
}

# Create figure
fig, axs = plt.subplots(nrows=1, ncols=len(logs)+1, figsize=(16, 10), sharey=True)

# Plot each log curve
for i, log in enumerate(logs):
    axs[i].plot(df_filtered[log], df_filtered['DEPTH_MD'], color=colors[i])
    axs[i].set_xlabel(xlabels[i], fontsize=17)
    axs[i].tick_params(axis='both', labelsize=15)
    axs[i].invert_yaxis()
    axs[i].grid(True)

# Plot lithology
lith = df_filtered['FORCE_2020_LITHOFACIES_LITHOLOGY']
depth = df_filtered['DEPTH_MD']
lith_colors = lith.map(lithology_colors).fillna('gray')

for i in range(len(df_filtered) - 1):
    axs[-1].fill_betweenx(
        [depth.iloc[i], depth.iloc[i + 1]],
        0, 1,
        color=lith_colors.iloc[i]
    )

axs[-1].set_xlabel("Lithology", fontsize=17)
axs[-1].set_xticks([])
axs[-1].tick_params(axis='y', labelsize=15)
axs[0].set_ylabel("Depth (m)", fontsize=18)
for ax in axs:
    ax.invert_yaxis()

# Legend
legend_patches = [
    mpatches.Patch(color=lithology_colors[code], label=label)
    for code, label in lithology_keys.items()
]
axs[-1].legend(handles=legend_patches, loc='upper center',
               bbox_to_anchor=(0.5, -0.05), fontsize=16)

# Final formatting
plt.tight_layout()

# Save the figure as PDF
plt.savefig("well_logs_lithology_plot.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
